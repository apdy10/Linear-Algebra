```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Algebraic and Geometric interpretations

```python
# We can create vector using lists and numpy arrays both

# 2- dimensional vector using list
v1 = [1,2]

              # converting list to numpy arrays

v1n = np.array(v1)

# 3-dimensional Vector using list

v2 = [1,2,3]

                # Converting list to numpy arrays

v2n = np.array(v2)


#Transpose - row to column(or vice versa)

v3 = np.transpose(v2)
print(v3)

                # Transpose - row to column(or vice versa) in case of numpy arrays

v3n = v2n.T
print(v3n)
```

**Geometrically vectors are defined by their length and directions**

```python
#plot them

plt.plot([0,v1[0]], [0,v1[1]])
plt.axis('equal')
plt.plot([-4,4],[0,0],'k--')
plt.plot([0,0],[-4,4],'k--')
plt.grid()
plt.axis((-4,4,-4,4))
plt.show()
```

Plot the 3D vector

```python
fig = plt.figure(figsize = plt.figaspect(1))
ax = fig.gca(projection = '3d')
ax.plot([0, v3[0]], [0,v3[1]], [0,v3[2]],linewidth = 3)


#making the plot look good

ax.plot([0,0],[0,0],[-4,4], 'k--')
ax.plot([0,0],[-4,4],[0,0], 'k--')
ax.plot([-4,4],[0,0],[-0,0], 'k--')
```
